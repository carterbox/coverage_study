```python
import xdesign as xd
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
p0 = xd.Softwood()
d = xd.sidebyside(p0, size=512)
plt.show()
```

```python
size = 512
labels = None
prop='mass_attenuation'
fig = plt.figure(figsize=(16, 9), dpi=600)

axis = fig.add_subplot(121, aspect='equal')
xd.plot_phantom(p0, axis=axis, labels=labels)
plt.grid('on')
axis.invert_yaxis()
axis.set_xticks(np.linspace(0, 1, 6, True))
axis.set_yticks(np.linspace(0, 1, 6, True))

axis = plt.subplot(122)
d = xd.discrete_phantom(p0, size, prop=prop)
plt.imshow(d, interpolation='none', cmap=plt.cm.inferno)
axis.set_xticks(np.linspace(0, size, 6, True))
axis.set_yticks(np.linspace(0, size, 6, True))

plt.tight_layout()
plt.savefig('figures/wood/first.png')
plt.show()
```

```python
size = 512
labels = None
prop='mass_attenuation'
fig = plt.figure(figsize=(16, 9), dpi=500)

for i in range(6):
    axis = fig.add_subplot(2,3,i+1, aspect='equal')
    p = xd.Softwood()
    d = xd.discrete_phantom(p, size, prop=prop)
    plt.imshow(d, interpolation='none', cmap=plt.cm.inferno)
    axis.set_xticks(np.linspace(0, size, 6, True))
    axis.set_yticks(np.linspace(0, size, 6, True))

plt.tight_layout()
plt.savefig('figures/wood/procedure.png')
plt.show()
```
